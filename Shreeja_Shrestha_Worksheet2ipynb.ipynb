```python
import pandas as pd
#Task1:Load the dataset
df=pd.read_csv("/content/drive/MyDrive/AI/medical_students_dataset.csv")
print(df.head())
```

```python
#Task2:Check Info and identify Columns with Missing Values
print("DataFrame Info:")
print(df.info())
#identify columns with missing values
missing_values=df.isnull().sum()
print("columns with missing values")
print(missing_values[missing_values>0])
```

```python
#Task3
# Fill missing values for numerical columns with mean
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Fill missing values for categorical columns with mode
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)


# Display the updated DataFrame info
print("\nUpdated DataFrame Info after Filling Missing Values:")
print(df.info())
```

```python
#TAsk4
#check for duplicate values
# Check for duplicate values
duplicates = df.duplicated().sum()
print(f"\nNumber of Duplicate Rows: {duplicates}")

# Drop duplicate rows
df = df.drop_duplicates()

# Display the DataFrame info after removing duplicates
print("\nDataFrame Info after Removing Duplicates:")
print(df.info())
```
