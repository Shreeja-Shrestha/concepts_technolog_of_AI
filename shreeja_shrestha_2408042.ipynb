
 Task- 3- Outlier Detection:
>



1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

 2. Define outliers using the 1.5 × IQR rule.


 3.Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
 in a

```python
import pandas as pd
import matplotlib.pyplot as plt

# List of South Asian countries
south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India","Maldives", "Nepal", "Pakistan", "Sri Lanka"]

# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/AI/WHR-2024-5CS037.csv')

# Filter the dataset to show only South Asian countries
south_asian_data = data[data['Country name'].isin(south_asian_countries)].copy()

# Calculate Q1 and Q3 for 'score'
Q1_score = south_asian_data['score'].quantile(0.25)
Q3_score = south_asian_data['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

# Calculate lower and upper bounds for 'score'
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

# Calculate Q1 and Q3 for 'Log GDP per capita'
Q1_gdp = south_asian_data['Log GDP per capita'].quantile(0.25)
Q3_gdp = south_asian_data['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# Calculate lower and upper bounds for 'Log GDP per capita'
lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Identify outliers in 'score'
score_outliers = south_asian_data[(south_asian_data['score'] < lower_bound_score) | (south_asian_data['score'] > upper_bound_score)]

# Identify outliers in 'Log GDP per capita'
gdp_outliers = south_asian_data[(south_asian_data['Log GDP per capita'] < lower_bound_gdp) | (south_asian_data['Log GDP per capita'] > upper_bound_gdp)]

# Combine the outliers from both metrics
combined_outliers = pd.concat([score_outliers, gdp_outliers]).drop_duplicates().reset_index(drop=True)
#Display the outliers
if combined_outliers.empty:
  print("No outlier countries based on score and GDP per capita.")
else:
  print("Outlier countries based on score and GDP per capita:")
  print(combined_outliers[['Country name', 'score', 'Log GDP per capita']])
print("hihi",score_outliers)

# Scatter plot with GDP per Capita on the x-axis and Score on the y-axis
plt.figure(figsize=(10, 6))

# Plotting the regular data points
plt.scatter(south_asian_data['Log GDP per capita'], south_asian_data['score'], label='Regular Data', color='blue', alpha=0.6)

# Highlighting the outliers
plt.scatter(combined_outliers['Log GDP per capita'], combined_outliers['score'], label='Outliers', color='red', edgecolor='black')

plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.title('Scatter Plot of GDP per Capita vs Happiness Score with Outliers Highlighted')
plt.legend()
plt.grid(True)

plt.show()
```

 2. Create scatter plots with trendlines for these metrics against the Score

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India","Maldives", "Nepal", "Pakistan", "Sri Lanka"]

data = pd.read_csv('/content/drive/MyDrive/AI/WHR-2024-5CS037.csv')

south_asian_data = data[data['Country name'].isin(south_asian_countries)].copy()

# Function to add a trendline
def add_trendline(x, y, ax):
    z = np.polyfit(x, y, 1)
    p = np.poly1d(z)
    ax.plot(x, p(x), "purple")

# Scatter plot with trendline for Freedom to Make Life Choices
plt.figure(figsize=(10, 6))
plt.scatter(south_asian_data['Freedom to make life choices'], south_asian_data['score'], color='blue', alpha=0.6)
plt.xlabel('Freedom to Make Life Choices')
plt.ylabel('Happiness Score')
plt.title('Freedom to Make Life Choices vs Happiness Score')

# Add trendline
add_trendline(south_asian_data['Freedom to make life choices'], south_asian_data['score'], plt)
plt.grid(True)
plt.show()

# Scatter plot with trendline for Generosity
plt.figure(figsize=(10, 6))
plt.scatter(south_asian_data['Generosity'], south_asian_data['score'], color='blue', alpha=0.6)
plt.xlabel('Generosity')
plt.ylabel('Happiness Score')
plt.title('Generosity vs Happiness Score')

# Add trendline
add_trendline(south_asian_data['Generosity'], south_asian_data['score'], plt)
plt.grid(True)
plt.show()
```

```python
import pandas as pd

south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India","Maldives", "Nepal", "Pakistan", "Sri Lanka"]

data = pd.read_csv('/content/drive/MyDrive/AI/WHR-2024-5CS037.csv')

south_asian_data = data[data['Country name'].isin(south_asian_countries)].copy()

# Add the GDP-Score Gap column
south_asian_data['GDP-Score Gap'] = south_asian_data['Log GDP per capita'] - south_asian_data['score']

# Display the updated DataFrame
print(south_asian_data[['Country name', 'Log GDP per capita', 'score', 'GDP-Score Gap']])
```

 2. Rank the South Asian countries by this gap in both ascending and descending order.


```python
# Rank the countries by GDP-Score Gap in ascending order
ascending_order = south_asian_data.sort_values(by='GDP-Score Gap',ascending=True)

# Rank the countries by GDP-Score Gap in descending order
descending_order = south_asian_data.sort_values(by='GDP-Score Gap', ascending=False)

# Display the results
print("South Asian countries ranked by GDP-Score Gap Ascending Order:")
print(ascending_order[['Country name', 'Log GDP per capita', 'score', 'GDP-Score Gap']])

print("South Asian countries ranked by GDP-Score Gap Descending Order:")
print(descending_order[['Country name', 'Log GDP per capita', 'score', 'GDP-Score Gap']])
```

```python
# range and CV for South Asia
south_asia_max = south_asian_data['score'].max()
south_asia_min = south_asian_data['score'].min()
south_asia_range = south_asia_max- south_asia_min
south_asia_mean = south_asian_data['score'].mean()
south_asia_std = south_asian_data['score'].std()
south_asia_cv = (south_asia_std / south_asia_mean) * 100

#  range and CV for Middle East
middle_east_max = middle_east_data['score'].max()
middle_east_min = middle_east_data['score'].min()
middle_east_range = middle_east_max - middle_east_min
middle_east_mean = middle_east_data['score'].mean()
middle_east_std = middle_east_data['score'].std()
middle_east_cv = (middle_east_std / middle_east_mean) * 100

# Print the results
print(f"South Asia - Range: {south_asia_range}, CV: {south_asia_cv:.2f}%")
print(f"Middle East - Range: {middle_east_range}, CV: {middle_east_cv:.2f}%")

# Determine which region has greater variability in happiness
greater_variability_region = "South Asia" if south_asia_cv > middle_east_cv else "Middle East"
print("Region with greater variability in happiness:", greater_variability_region)

```
